#### 3. Explain the following terms:
##### a) The difference in np.array, np.asarray and np.asanyarray.
##### b) The difference between Deep copy and shallow copy.

```python
# np.array: Always creates a new array.
a = [1, 2, 3]
b = np.array(a)
print(b)

# np.asarray: Converts input to an array, avoids copying if the input is already an ndarray.
a = np.array([1, 2, 3])
b = np.asarray(a)
print(b is a)

# np.asanyarray: Similar to np.asarray, but preserves ndarray subclasses.
a = np.matrix([[1, 2], [3, 4]])
b = np.asanyarray(a)
print(b is a)
```

```python
# Shallow Copy: Copies the structure, but not the elements themselves if they are mutable objects (e.g., lists within lists).
original = [1, 2, [3, 4]]
shallow = copy.copy(original)
shallow[2][0] = 'a'
print("Shallow Copy : ")
print("Original :", original)
print("Shallow :", shallow)

# Deep Copy: Copies both the structure and all elements recursively, ensuring complete independence from the original.
print("\nDeep Copy : ")
original = [1, 2, [3, 4]]
deep = copy.deepcopy(original)
deep[2][0] = 'a'
print("Original :", original)
print("Deep :", shallow)
```

#### 14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
##### a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
##### b) Label the x-axis as 'Date'.
##### c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
np.random.seed(0)
date_range = pd.date_range(start='2024-01-01', periods=100, freq='D')
temperature = np.random.uniform(low=15, high=30, size=len(date_range))
humidity = np.random.uniform(low=30, high=90, size=len(date_range))
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})
df.set_index('Date', inplace=True)

fig, ax1 = plt.subplots(figsize=(12, 5))

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df.index, df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')
ax1.grid(True)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)
ax2.plot(df.index, df['Humidity'], color=color, label='Humidity')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Temperature and Humidity Over Time')
fig.tight_layout()
plt.show()
```

#### 15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
##### a) Plot a histogram of the data with 30 bins.
##### b) Overlay a line plot representing the normal distribution's probability density function (PDF).
##### c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
##### d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
data = np.random.randn(1000)
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

mu, std = np.mean(data), np.std(data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram with Normal Distribution PDF Overlay')
plt.legend()

plt.grid()
plt.show()
```

#### 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
np.random.seed(0)
x = np.random.uniform(-10, 10, size=100)
y = np.random.uniform(-10, 10, size=100)

df = pd.DataFrame({
    'x': x,
    'y': y
})

def determine_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant 1'
    elif x < 0 and y > 0:
        return 'Quadrant 2'
    elif x < 0 and y < 0:
        return 'Quadrant 3'
    elif x > 0 and y < 0:
        return 'Quadrant 4'
    else:
        return 'On Axis'

df['Quadrant'] = [determine_quadrant(xi, yi) for xi, yi in zip(df['x'], df['y'])]
sns.scatterplot(x='x', y='y', data=df, hue='Quadrant', palette='Set1', s=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.grid()
plt.show()
```

#### 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

#### 19. Usig Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

```python
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
colors = []
values = [np.random.randint(1, 100) for _ in categories]

source = ColumnDataSource(data=dict(categories=categories, values=values))
p = figure(x_range=categories, title="Random Categorical Bar Chart", toolbar_location=None, tools="", tooltips="@categories: @values")
p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories",
       line_color='white', fill_color=factor_cmap('categories', palette=Category10[len(categories)], factors=categories))
hover = HoverTool()
hover.tooltips = [
    ("Category", "@categories"),
    ("Value", "@values")
]
p.add_tools(hover)

p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```
